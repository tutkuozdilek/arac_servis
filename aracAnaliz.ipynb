```python
ad["tarih"] = pd.to_datetime(ad["tarih"])
```

```python
ad2 = ad[["tarih","parca_fiyat"]]
```

```python
ad2=ad2.rename(columns={"tarih":"ds","parca_fiyat":"y"})
```

```python
ad.info()
```

```python
m = Prophet(daily_seasonality=True)
m.fit(ad2)
```

```python
future = m.make_future_dataframe(periods=30)
future.tail(30)
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(30)
```
